```python
# Построить график
from sympy.core import symbol
from sympy.plotting import plot
from sympy import Symbol, solve, diff, sin, cos, evalf, solveset, S, pprint
x = Symbol('x')
f = sin(x)**2 - cos(x)**2
plot(f)
```

```python
# Определить корни
print('Корни уравнения:')
ans = solve(f,x)
print(ans)
for i in ans:
  print(i.evalf())
```

```python
# Определить промежутки, на котором f > 0
pprint(solveset(f > 0, x, domain = S.Reals), use_unicode = True)
```

```python
# Определить промежутки, на котором f < 0
#pprint(solveset(f < 0, x, domain = S.Reals), use_unicode = True)
solve(f < 0, x)
```

```python
# Вычислить вершину
diff_f = diff(f, x)
print(diff_f)
solve(diff_f, x)
```

```python
#Найти интервалы, на которых функция возрастает
pprint(solveset(diff_f > 0, x, domain = S.Reals), use_unicode = True)
solve(diff_f > 0, x)
```
